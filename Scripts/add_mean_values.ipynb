```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from math import exp
```

```python
data2 = pd.read_csv('/Users/alekseyvalouev/Desktop/FireData/CSV/final_table_staggered.csv')
```

```python
data2 = data2.reset_index(drop=True)

vpd = data2.groupby("newID", as_index=False).VPD.rolling(12, closed='left').mean()

tmax = data2.groupby("newID", as_index=False).TMAX.rolling(12, closed='left').mean()

ppt = data2.groupby("newID", as_index=False).PPT.rolling(12, closed='left').mean()
```

```python
vpd.rename(columns = {"VPD" : "12MoMeanVPD", "newID" : "ID"}, inplace = True)

t = pd.concat([data2, vpd], axis=1, join="outer")
assert(t['newID'].equals(t['ID']))
t = t.drop(['ID'], axis=1)
```

```python
tmax.rename(columns = {"TMAX" : "12MoMeanTMAX", "newID" : "ID"}, inplace = True)

t = pd.concat([t, tmax], axis=1, join="outer")
assert(t['newID'].equals(t['ID']))
t = t.drop(['ID'], axis=1)
```

```python
ppt.rename(columns = {"PPT" : "12MoMeanPPT", "newID" : "ID"}, inplace = True)

t = pd.concat([t, ppt], axis=1, join="outer")
assert(t['newID'].equals(t['ID']))
t = t.drop(['ID'], axis=1)
```

```python
cols = t.columns.tolist()
cols = cols[:25] + cols[-3:] + cols[25:-3]
t = t[cols]
```
