```python
import pandas as pd
```

```python
data2 = pd.read_csv("/Users/alekseyvalouev/Desktop/FireData/CSV/final_table_perm_added.csv")

#data2 = data2.dropna(axis=0, subset=['fut1fire'])
```

```python
data2 = data2.reset_index(drop=True)

t = data2
tmp = t.groupby("newID", as_index=False).Fire.rolling(36, closed='left').sum()
tmp2 = t.groupby("newID", as_index=False).Fire.rolling(12, closed='left').sum()
tmp.rename(columns = {"Fire" : "36MoFires", "newID" : "ID"}, inplace = True)
tmp2.rename(columns = {"Fire" : "12MoFires", "newID" : "ID"}, inplace = True)
tmp = pd.DataFrame(tmp)
tmp = tmp.reset_index(drop=True)
tmp2 = pd.DataFrame(tmp2)
tmp2 = tmp2.reset_index(drop=True)
t = pd.concat([t, tmp], axis=1, join="outer")
assert(t['newID'].equals(t['ID']))
t = t.drop(['ID'], axis=1)

t = pd.concat([t, tmp2], axis=1, join="outer")
assert(t['newID'].equals(t['ID']))
t = t.drop(['ID'], axis=1)

t['Diff'] = t['36MoFires'].sub(t['12MoFires'], axis = 0)

cols = t.columns.tolist()
cols = cols[:13] + cols[-3:] + cols[13:-3]
t = t[cols]
```
