```python
import pandas as pd
```

```python
table2 = pd.read_csv("/Users/alekseyvalouev/Desktop/FireData/CSV/finaltable2.csv")
```

```python
table2
```

```python
og_cols = list(table2.columns)
vpd1 = og_cols.index('202108_vpd')
vpd2 = og_cols.index('199901_vpd')
vpd3 = og_cols.index('202107_vpd') + 1
og_cols = og_cols[:vpd1] + og_cols[vpd2:vpd3] + og_cols[vpd1:vpd2] + og_cols[vpd3:]
tma1 = og_cols.index('202108_tma')
tma2 = og_cols.index('199901_tma')
tma3 = og_cols.index('202107_tma') + 1
og_cols = og_cols[:tma1] + og_cols[tma2:tma3] + og_cols[tma1:tma2] + og_cols[tma3:]
ppt1 = og_cols.index('202108_ppt')
ppt2 = og_cols.index('199901_ppt')
ppt3 = og_cols.index('202107_ppt') + 1
og_cols = og_cols[:ppt1] + og_cols[ppt2:ppt3] + og_cols[ppt1:ppt2] + og_cols[ppt3:]
cols = og_cols
```

```python
table2 = table2[cols]
```

```python
sp1 = list(cols[7:int(7+831/3)])
sp2 = list(cols[int(7+831/3):int(7+831*2/3)])
sp3 = list(cols[int(7+831*2/3):int(7+831)])
```

```python
ordered = table2.groupby('newID').sum().reset_index()
```

```python
vpd = ordered.melt(id_vars=['newID'], 
value_vars=sp1, var_name='YEAR', value_name='VPD')
vpd = vpd.sort_values(['newID', 'YEAR'], ascending=True)
vpd = vpd.reset_index(drop=True)
```

```python
ppt = ordered.melt(id_vars=['newID'], 
value_vars=sp2, var_name='YEAR', value_name='PPT')
ppt = ppt.sort_values(['newID', 'YEAR'], ascending=True)
ppt = ppt.reset_index(drop=True)
```

```python
tmax = ordered.melt(id_vars=['newID'], 
value_vars=sp3, var_name='YEAR', value_name='TMAX')
tmax = tmax.sort_values(['newID', 'YEAR'], ascending=True)
tmax = tmax.reset_index(drop=True)
```

```python
result = pd.concat([vpd, ppt, tmax], axis=1)
```

```python
result = result.loc[:,~result.columns.duplicated()]
```

```python
def truncate_cat(x):
    return x[:-4]
```

```python
result.YEAR = result.YEAR.apply(truncate_cat)
```

```python
result.rename(columns = {"YEAR" : "Date"}, inplace = True)
```

```python
result['Date'] = pd.to_datetime(result['Date'], format='%Y%m')
```
