```python
table1 = table1.reset_index(drop=True)

count = 0

for index, row in table1.iterrows():
    newid = row["newID"]
    date = row["date_formatted"]
    
    if (merged.loc[(merged['newID'] == newid) & (merged['YYYYMM'] == date)].any()["Fire"] != 1):
        print(str(newid) + "_" + str(date))
        count+=1
print(count)
```
